```python
import zipfile
import glob
import os
```

```python
zip_dir = 'Datasets/Zips'
output_dir = 'Datasets'

for zip_file in os.listdir(zip_dir):
    if zip_file.endswith('.zip'):
        file_path = os.path.join(zip_dir, zip_file)
        with zipfile.ZipFile(file_path, 'r') as zip_ref: zip_ref.extractall(output_dir)
```

```python
for folder in ['Datasets/train/labels', 'Datasets/valid/labels', 'Datasets/test/labels']:
    label_files = glob.glob(os.path.join(folder, '*.txt'))
    
    for file_path in label_files:
        
        with open(file_path, 'r') as f:
            lines = f.readlines()
        
        new_lines = []
        for line in lines:
            parts = line.strip().split()
            if len(parts) > 0:
                parts[0] = '0'  # Convert to string before joining
                new_lines.append(' '.join(parts) + '\n')
        
        with open(file_path, 'w') as f: f.writelines(new_lines)
```
